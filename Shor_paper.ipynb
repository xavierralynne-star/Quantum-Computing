```python
def check_if_power(N):
    b=2
    while (2**b) <= N:
        a = 1
        c = N
        while (c-a) >= 2:
            m = int((a+c)/2)

            if (m**b) < (N+1):
                m = int((m**b))
            else:
                p = int(N+1)

            if int(p) == int(N):
                print('N adalah {0}^{1}'.format(int(m),int(b)))
                return True
            
            if p<N:
                a = int(m)
            else:
                c = int(m)
        b = b+1
    return False
```
